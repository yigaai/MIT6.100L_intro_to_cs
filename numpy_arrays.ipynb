```python
import random
import matplotlib.pyplot as plt

# Function to simulate decision-making
def decision_making(threshold, evidence_per_trial):
    accumulated_evidence = 0
    trials = 0
    
    while accumulated_evidence < threshold:
        # Simulate gathering evidence (random positive or negative)
        accumulated_evidence += random.choice([1, -1]) * evidence_per_trial
        trials += 1
        
    return trials

# Run simulations for decision-making
threshold = 10  # Threshold of evidence to make a decision
evidence_per_trial = 1  # Amount of evidence gathered per trial

# Simulate decision-making for 100 subjects
num_simulations = 100
results = [decision_making(threshold, evidence_per_trial) for _ in range(num_simulations)]

# Plot the results (how many trials it took to reach the threshold)
plt.hist(results, bins=15, color='blue', alpha=0.7)
plt.title("Decision-Making Model: Number of Trials to Reach Threshold")
plt.xlabel("Number of Trials")
plt.ylabel("Frequency")
plt.show()
```
